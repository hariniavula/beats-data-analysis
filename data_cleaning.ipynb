```python
import pandas as pd
```

```python
file_path = 'Reviews-update.csv'
df = pd.read_csv(file_path, quotechar='"')
df
```

```python
# Convert data types if necessary
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Check the updated data types
print(df.dtypes)
```

```python
# Check for duplicates
duplicates = df.duplicated()
print(duplicates.sum())
```

```python
null_ratings = df['rating'].isnull().sum()
print(f"Number of null ratings: {null_ratings}")
```

```python
null_values = df.isnull().sum()
print(null_values)
##leaving these null values as timestamp, content, and product-id are not necessary for calculations
## most important values are rating and title
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the rating column
Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['rating'] < lower_bound) | (df['rating'] > upper_bound)]
print(f"Detected outliers in 'rating':\n{outliers}")
```
